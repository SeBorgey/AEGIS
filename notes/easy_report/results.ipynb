```python
import pandas as pd
from IPython.display import display, HTML

try:
    df = pd.read_csv('results.csv', sep=';')
    df.fillna({'заметки_1': '', 'заметки_2': '', 'заметки_3': ''}, inplace=True)
    print("Данные успешно загружены.")
    display(df.head())
except FileNotFoundError:
    print("Ошибка: Файл 'app_stats.csv' не найден. Убедитесь, что он находится в той же директории, что и ноутбук.")
```

```python
df['Всего успешных попыток'] = df['успех_1'] + df['успех_2'] + df['успех_3']
df['Всего попыток'] = df['попыток_1'] + df['попыток_2'] + df['попыток_3']
df['Процент успеха'] = (df['Всего успешных попыток'] / df['Всего попыток'] * 100).round(2)

summary_df = df[['приложение', 'Процент успеха', 'Всего успешных попыток', 'Всего попыток']].copy()
summary_df['Процент успеха'] = summary_df['Процент успеха'].astype(str) + '%'

display(HTML('<h2>Сводная таблица успехов</h2>'))
display(summary_df.style.set_properties(**{'text-align': 'left'}).set_table_styles([dict(selector='th', props=[('text-align', 'left')])]))
```

```python
detailed_df = df[['приложение', 'успех_1', 'заметки_1', 'успех_2', 'заметки_2', 'успех_3', 'заметки_3']].copy()

def style_success(val):
    if val == 1:
        return 'color: green; font-weight: bold;'
    elif val == 0:
        return 'color: red; font-weight: bold;'
    return ''

styled_detailed_df = detailed_df.style.applymap(style_success, subset=['успех_1', 'успех_2', 'успех_3'])

detailed_df['успех_1'] = detailed_df['успех_1'].apply(lambda x: 'Да' if x == 1 else 'Нет')
detailed_df['успех_2'] = detailed_df['успех_2'].apply(lambda x: 'Да' if x == 1 else 'Нет')
detailed_df['успех_3'] = detailed_df['успех_3'].apply(lambda x: 'Да' if x == 1 else 'Нет')


display(HTML('<h2>Детальная таблица по попыткам</h2>'))
display(styled_detailed_df.set_properties(**{'text-align': 'left', 'vertical-align': 'top'}).set_table_styles([dict(selector='th', props=[('text-align', 'left')])]))
```

```python
from IPython.display import display, HTML

# Формирование HTML для карточек с заметками
# Добавлены стили для черного цвета шрифта
notes_html = "<h2 style='color: white;'>Заметки по приложениям</h2><div style='display: grid; grid-template-columns: repeat(auto-fit, minmax(300px, 1fr)); gap: 20px;'>"

for index, row in df.iterrows():
    notes_list = []
    # Собираем все непустые заметки
    if row['заметки_1']:
        notes_list.append(f"<li>{row['заметки_1']}</li>")
    if row['заметки_2']:
        notes_list.append(f"<li>{row['заметки_2']}</li>")
    if row['заметки_3']:
        notes_list.append(f"<li>{row['заметки_3']}</li>")

    # Если для приложения есть заметки, создаем карточку
    if notes_list:
        notes_html += f"""
        <div style='background: #fff; border: 1px solid #ddd; border-radius: 5px; padding: 20px; box-shadow: 0 2px 5px rgba(0,0,0,0.1);'>
            <h3 style='margin-top: 0; color: #0275d8;'>{row['приложение']}</h3>
            <ul style='padding-left: 20px; list-style: disc; color: black;'>
                {''.join(notes_list)}
            </ul>
        </div>
        """
notes_html += "</div>"

display(HTML(notes_html))
```
